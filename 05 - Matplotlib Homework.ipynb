```python
import os
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import time
import json
import seaborn
from scipy.stats import ttest_ind
import math
```

```python
city_data = "city_data.csv"
city_dataset_1 = pd.read_csv(city_data)
city_dataset_1.head()
```

```python
ride_data = "ride_data.csv"
ride_dataset_1 = pd.read_csv(ride_data)
ride_dataset_1.head()
```

```python
city_and_ride = pd.merge(city_dataset_1, ride_dataset_1, how="left", on=["city", "city"])
city_and_ride.head()
```

```python
#Average Fare ($) Per City
# search first?
av_fare = city_and_ride.groupby(["city"])["fare"].mean()
av_fare 
```

```python
# * Total Number of Rides Per City
urban_ride_count = city_and_ride.groupby(["city"])["ride_id"].count()
urban_ride_count
```

```python
# * Total Number of Drivers Per City
drivers_per_city = city_and_ride.groupby(["city"])["driver_count"].count()
drivers_per_city 
```

```python
#* City Type (Urban, Suburban, Rural)
urban_cities = city_and_ride[city_and_ride["type"] == "Urban"]
suburban_cities = city_and_ride[city_and_ride["type"] == "Suburban"]
rural_cities = city_and_ride[city_and_ride["type"] == "Rural"]
```
